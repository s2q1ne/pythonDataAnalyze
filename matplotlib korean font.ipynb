```python
%matplotlib inline
#패키지 가져오기
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm
mpl.rcParams['axes.unicode_minus'] = False
```

```python
#임의로 데이터 만들기
import numpy as np
data = np.random.randint(-100,100,50).cumsum()
data
```

```python
#그래프 그리기(한글 깨짐)
plt.plot(range(50), data, 'r')
mpl.rcParams['axes.unicode_minus'] = False
plt.title("시간별 가격 추이")
plt.ylabel("주식 가격")
plt.xlabel("시간(분)")
```

```python
#패키지 matplotlib 정보확인
print('버전: ', mpl.__version__)
print('설치 위치: ', mpl.__file__)
print('설정 위치: ', mpl.get_configdir())
print('캐시 위치: ', mpl.get_cachedir())
```

```python
#터미널 위치 이동
print('설정파일 위치: ', mpl.matplotlib_fname())
```

```python
# fname 옵션 사용
path = 'C:\\WINDOWS\\Fonts\\NanumGothic.ttf'
fontprop = fm.FontProperties(fname=path, size=18)

plt.plot(range(50), data, 'r')
plt.title("시간별 가격 추이", fontProperties=fontprop )
plt.ylabel("주식 가격", fontProperties=fontprop)
plt.xlabel("시간(분)", fontProperties=fontprop)
```

```python
#기본 설정
import matplotlib.pyplot as plt

#size, family
print("# 설정되어 있는 폰트 사이즈", plt.rcParams['font.size'])
print("# 설정되어 있는 폰트 글꼴", plt.rcParams['font.family'])
```

```python
print('serif 있는 폰트', plt.rcParams['font.serif'])
print('sans-serif 폰트', plt.rcParams['font.sans-serif'])
print('monospace 폰트', plt.rcParams['font.monospace'])
```

```python
plt.rcParams['font.family'] = 'NanumSquareRound'
plt.rcParams['font.size'] = 20
plt.rcParams['figure.figsize'] = (14,4)
```

```python
plt.plot(range(50), data, 'r')
plt.title('시간별 가격 추이')
plt.ylabel('주식 가격')
plt.xlabel('시간(분)')
plt.style.use('seaborn-pastel')
plt.show()
```

```python
path = 'C:\\WINDOWS\\Fonts\\NanumGothic.ttf'
font_name = fm.FontProperties(fname=path, size=50).get_name()
plt.rc('font', family=font_name)

fig, ax = plt.subplots()
ax.plot(data)
ax.set_title('시간별 가격 추이')
plt.ylabel('주식 가격')
plt.xlabel('시간(분)')
plt.style.use('ggplot')
plt.show()
```
